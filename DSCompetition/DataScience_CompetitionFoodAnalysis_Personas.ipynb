```python
m_food_insecure= train_data[train_data["FOOD_INSECURE"] == 1]
m_food_insecure
```

```python
m_mask= m_food_insecure['PESEX'] == 2
female_fi_df= m_food_insecure[m_mask]
male_fi_df= m_food_insecure[~m_mask]
```

```python
m_mask= female_fi_df['PEMARITL'] > 2
m_single_female= female_fi_df[m_mask]
m_mask= m_single_female['PRCHLD'] > 0
m_single_moms_df= m_single_female[m_mask]
m_single_moms_df
```

```python
def execute_expectation_maximization_gmm(k, working_df_arr):
    rng= RandomState(42)

    gmm = GMM(n_components=k, covariance_type='full', init_params='wmc', n_iter=200, random_state=rng)
    gmm.fit(working_df_arr)
    
    m_means= gmm.means_
    m_labels= gmm.predict(working_df_arr)
    return (m_means, m_labels, gmm)
```

```python
# process single moms
working_df= m_single_moms_df[final_selected_predictors]
working_df.reset_index(drop=True)
working_df_arr= working_df.values

# use PCA to identify clusters visually
X= working_df_arr
pca = PCA(n_components=3, whiten=True).fit(X)
X_pca = pca.transform(X)
xs= X_pca[:,0]
ys= X_pca[:,1]
zs= X_pca[:,2]

fig,ax= plt.subplots(figsize=(10,5))

ax.scatter(xs, ys, c='b', marker='o')

ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
```
