```python
class Model:
    def __init__(self, train_x, train_x_labels, fields= None):
        self._mFields= fields
        self._mModel= self._init_model(train_x, train_x_labels, fields)
    
    def _init_model(self, train_x, train_x_labels, fields):
        raise NotImplementedError
    
    def evaluate(self, test_y, test_y_labels):
        return self._do_evaluate(self._mModel, test_y, test_y_labels, self._mFields)
    
    def _do_evaluate(self, model, test_y, test_y_labels, fields):
        raise NotImplementedError
```

```python
class LogisticRegressionModel(Model):
    def __init__(self, train_x, train_x_labels, fields= None):
        Model.__init__(self, train_x, train_x_labels, fields)
    
    def _init_model(self, train_x, train_x_labels, fields):
        rng = RandomState(45)
        m_model= LogisticRegression(random_state=rng)
        if fields is None:
            m_model.fit(train_x, train_x_labels)
        else:
            m_model.fit(train_x[fields], train_x_labels)
        return m_model
    
    def _do_evaluate(self, model, test_y, test_y_labels, fields):
        return model.score(test_y[fields],test_y_labels) if fields else model.score(test_y, test_y_labels)
```

```python
def new_logistic_regression_model(train_x, train_x_labels, feature_names):
    return LogisticRegressionModel(train_x, train_x_labels, feature_names)
```

```python
def select_best_feature(f_new_model, feature_names, test_columns, train_x, train_x_labels, test_y, test_y_labels):
    m_best_prediction= -1.
    m_best_feature= None
    #
    for c in test_columns:
        try:
            feature_names.append(c)
            m_model= f_new_model(train_x, train_x_labels, feature_names)
            c_prediction= m_model.evaluate(test_y, test_y_labels)
            if (c_prediction > m_best_prediction):
                m_best_feature= c
                m_best_prediction= c_prediction
        finally:
            feature_names.pop()
    if m_best_feature is None:
        raise RuntimeError("m_best_feature should not be null")
    feature_names.append(m_best_feature)
    test_columns.remove(m_best_feature)
    return (m_best_feature, m_best_prediction)
```

```python
num_desired_features= 60
print "Begin selection of features"
for i in xrange(num_desired_features):
    print "selecting feature ", i
    start_time = int(round(time_.time() * 1000))
    m_result= select_best_feature(new_logistic_regression_model, feature_names_out, test_columns, x_train, x_train_labels, x_validation, x_validation_labels)
    stop_time = int(round(time_.time() * 1000))
    print "{} found in {}ms".format(m_result, stop_time - start_time)

print m_result
print feature_names_out
```
